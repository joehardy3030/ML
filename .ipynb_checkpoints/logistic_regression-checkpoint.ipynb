```python
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.linear_model import LogisticRegression
from pyML.plot_decision_regions import plot_decision_regions
from pyML.pymldata import iris_data_for_training
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)
```

```python
plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.axhspan(0.0, 1.0, facecolor='1.0', alpha=1.0, ls='dotted')
plt.yticks([0.0, 0.5, 1.0])
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
```

```python
plt.show()
```

```python
lr = LogisticRegression(C=1000.0, random_state=0)
X_train, X_test, y_train, y_test, X_train_std, X_test_std, X_combined_std, y_combined = iris_data_for_training()
```

```python
lr.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, 
                     y_combined, classifier=lr, 
                     test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

```python
lr.predict_proba(X_test_std)
```

```python
weights, params = [], []
for c in np.arange(-5,5):
    lr = LogisticRegression(C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10**c)
weights = np.array(weights)
plt.plot(params, weights[:, 0],
        label='petal length')
plt.plot(params, weights[:, 1], linestyle='--',
        label='petal width')
plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
plt.show()
```
