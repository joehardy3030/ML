```python
def get_share_values():
    price_array = []
    date_array = []
    for symbol in portfolio_df['symbol']:
        print symbol
        url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=%s&apikey=M3M3L3CIF19PI4Z4' % symbol
        print url
        r = requests.get(url)
        json_data = r.json()
        json_time_series = json_data['Time Series (Daily)']
        df = pd.DataFrame(json_time_series)
        df = df.T
        close_value = df['4. close']
        price_array = np.append(price_array,close_value.iloc[0]) 
        date_array = np.append(date_array,close_value.index[0]) 
    return price_array 
```

```python
price_array = []
date_array = []
for symbol in portfolio_df['symbol']:
    print symbol
    url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=%s&apikey=M3M3L3CIF19PI4Z4' % symbol
    print url
    r = requests.get(url)
    json_data = r.json()
    json_time_series = json_data['Time Series (Daily)']
    df = pd.DataFrame(json_time_series)
    df = df.T
    close_value = df['4. close']
    price_array = np.append(price_array,close_value.iloc[99]) 
    date_array = np.append(date_array,close_value.index[99]) 
    print price_array
    
```

```python
portfolio_df = portfolio_df.assign(price=pd.to_numeric(price_array))
```

```python
value_array = portfolio_df['price'] * portfolio_df['shares']
```

```python
portfolio_df = portfolio_df.assign(value=value_array)
```

```python
portfolio_df = portfolio_df.assign(date=date_array)
```

```python
import sqlite3
conn = sqlite3.connect('example.db')
```

```python
c = conn.cursor()
# Create table
c.execute('''CREATE TABLE stocks
             (date text, trans text, symbol text, qty real, price real)''')

# Insert a row of data
c.execute("INSERT INTO stocks VALUES ('2006-01-05','BUY','RHAT',100,35.14)")

# Save (commit) the changes
conn.commit()

# We can also close the connection if we are done with it.
# Just be sure any changes have been committed or they will be lost.
conn.close()
```
