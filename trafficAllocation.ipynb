```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import string as str
import seaborn as sns
import os
```

```python
df = pd.read_csv('./data/CW34_leads.csv')
```

```python
def make_date_cols(df):
    df['lead_creation_datetime'] = pd.to_datetime(pd.Series(df['Lead Creation Datetime']))
    df['created_date'] = pd.to_datetime(pd.Series(df['Created Date']))
    df['datetime_of_first_call'] = pd.to_datetime(pd.Series(df['DateTime Of First Call']))
    df['converted_date'] = pd.to_datetime(pd.Series(df['Converted Date']))
    return(df)
```

```python
df = make_date_cols(df)
```

```python
df['Landing URL'].str.contains('m.hear.com', regex=False)
```

```python
df['unbounce_primax'] = df['Landing URL'].str.contains('m.hear.com', regex=False)
```

```python
df['static_primax'] = df['Landing URL'].str.contains('m.cdn.hear.com/product-primax', regex=False)
```

```python
df['static_german'] = df['Landing URL'].str.contains('m.cdn.hear.com/product-primax_german-quality', regex=False)
```

```python
df[:10]
```

```python
df['static_primax_unique'] = (df['static_primax'] != df['static_german'])
```

```python
df_no_SMS = df[df['Marketing Partner'] != "SMS"]
```

```python
df_no_SMS.shape
```

```python
ct_LPs_devices = pd.crosstab(index=df_no_SMS['User Device'], columns = [df_no_SMS['static_primax_unique'], 
                                                                        df_no_SMS['static_german'], 
                                                                        df_no_SMS['unbounce_primax']])
```
